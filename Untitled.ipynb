```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt' 
```

```python
frozen_model ='frozen_inference_graph.pb' 
```

```python
model =cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLabels = []
filename = 'labels.txt'
with open(filename, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320, 320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5, 127.5, 127.5))
model.setInputSwapRB(True)
```

```python
img = cv2.imread('kids_playing.jpeg')
plt.imshow(img)
```

```python
classIndex, confidence, bbox = model.detect(img, confThreshold = 0.5)
```

```python
print(classIndex)
print(confidence)
print(bbox)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for classIndex, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img, boxes, (255, 0, 0), 2)
    cv2.putText(img, classLabels[classIndex-1], (boxes[0]+10, boxes[0]+40), font, fontScale = font_scale, color=(0, 255, 0), thickness = 3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cant open the video')
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    classIndex, confidence, bbox = model.detect(frame, confThreshold = 0.55)
    
    print(classIndex)
    
    if(len(classIndex) != 0):
        for classIndex, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox):
            if(classIndex <= 80):
                cv2.rectangle(frame, boxes, (255, 0, 0), 5)
                cv2.putText(frame, classLabels[classIndex-1], (boxes[0]+10, boxes[0]+40), font, fontScale = font_scale, color=(0, 255, 0), thickness = 3)
            
    cv2.imshow('object Detection', frame)
    
    if(cv2.waitKey(2) & 0xff == ord('q')):
        break

cap.release()
cv2.destroyAllWindows()
```
